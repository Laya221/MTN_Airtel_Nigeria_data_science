```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('filtered_comments_mymtn_cleaned.csv')

# Convert the commentTimestamp to datetime
df['commentTimestamp'] = pd.to_datetime(df['commentTimestamp'])

# Extract the month from the timestamp
df['month'] = df['commentTimestamp'].dt.to_period('M')

# Group by month and sentiment, then count the occurrences
sentiment_counts = df.groupby(['month', 'sentiment/finalClassification']).size().unstack(fill_value=0)

# Convert Period to string for plotting
sentiment_counts.index = sentiment_counts.index.astype(str)

# Define a color scheme
colors = {'positive': 'green', 'neutral': 'blue', 'negative': 'red'}

# Plot the data
ax = sentiment_counts.plot(kind='bar', stacked=True, figsize=(14, 8), color=[colors[col] for col in sentiment_counts.columns])

# Add annotations
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    if height > 0:  # Only annotate bars with height > 0
        ax.annotate(f'{int(height)}', (x + width / 2, y + height / 2), ha='center', va='center', fontsize=9, color='white')

# Add gridlines
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Improve title and labels
plt.title('Monthly Sentiment Distribution of MyMTN Comments', fontsize=16, pad=20)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Comments', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust legend position and style
plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('filtered_comments_mymtn.csv')

# Define keywords for each subcategory
app_stability_keywords = [
    "crash", "unresponsive", "session timeout", "login issue", "freeze", "lag", "slow"
]

account_sync_keywords = [
    "transaction delay", "balance discrepancy", "missing transaction", "sync issue", "balance not updating"
]

payment_processing_keywords = [
    "failed transaction", "temporary deduction", "duplicate charge", "refund delay", "payment issue"
]

# Function to categorize comments based on keywords
def categorize_comment(comment):
    comment = comment.lower()
    if any(keyword in comment for keyword in app_stability_keywords):
        return "App Stability"
    elif any(keyword in comment for keyword in account_sync_keywords):
        return "Account Synchronization"
    elif any(keyword in comment for keyword in payment_processing_keywords):
        return "Payment Processing"
    else:
        return None

# Apply the categorization function to the comments
df['category'] = df['commentText'].apply(categorize_comment)

# Filter out comments that are not related to MyMTN technical issues
technical_issues_df = df[df['category'].notna()]

# Group comments by category and include comment author
grouped_comments = technical_issues_df.groupby('category').apply(
    lambda x: x[['commentAuthor', 'commentText']].to_dict('records')
).reset_index(name='comments')

# Display the results
for index, row in grouped_comments.iterrows():
    print(f"Category: {row['category']}")
    print("Comments:")
    for comment in row['comments']:
        print(f"- Author: {comment['commentAuthor']}")
        print(f"  Comment: {comment['commentText']}")
    print("\n" + "="*50 + "\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate

# Load dataset
file_path = "final_mtn.csv"
df = pd.read_csv(file_path)

# Ensure text data is treated as string and lowercase for consistency
df["commentText"] = df["commentText"].astype(str).str.lower()

# Define category keywords
categories = {
    "App Stability": [
        "crash", "unresponsive", "freeze", "lag", "slow", "session timeout", "loading issue"
    ],
    "Account Synchronization": [
        "delay", "not updated", "incorrect balance", "missing transaction", "sync issue"
    ],
    "Payment Processing": [
        "failed transaction", "deduction", "duplicate charge", "refund", "receipt issue"
    ]
}

# Function to categorize comments
def categorize_comment(text):
    for category, keywords in categories.items():
        if any(keyword in text for keyword in keywords):
            return category
    return "Other"

# Apply categorization
df["category"] = df["commentText"].apply(categorize_comment)

# Filter for MyMTN-related comments
mymtn_keywords = ["MyMTN", "MyMTN app","app","application","mtn application","APP","App","mymtn","myMTN","Application","MTN application", "mtn app", "mymtn ng", "mtn nigeria app"]
df_mymtn = df[df["commentText"].str.contains("|".join(mymtn_keywords), na=False, regex=True)]

# Select relevant columns
df_mymtn_filtered = df_mymtn[["commentText", "commentAuthor", "commentLikesCount", "category"]]

# Save categorized MyMTN comments
df_mymtn_filtered.to_csv("mymtn_categorized_comments.csv", index=False)

# Count comments per category for visualization
category_counts = df_mymtn_filtered["category"].value_counts()

# === PLOTTING ===

# Bar chart for category distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="coolwarm")
plt.title("Distribution of Comments by Category", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Number of Comments", fontsize=12)
plt.xticks(rotation=20)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Pie chart for category distribution
plt.figure(figsize=(6, 6))
category_counts.plot.pie(autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Percentage of Comments by Category", fontsize=14)
plt.ylabel("")  # Hide y-label
plt.show()

# === DISPLAY TABLES WITH ENHANCED FORMATTING ===
for category in categories.keys():
    df_category = df_mymtn_filtered[df_mymtn_filtered["category"] == category].sort_values(by="commentLikesCount", ascending=False)
    
    if not df_category.empty:
        print(f"\n=== {category} ===")
        print(tabulate(df_category, headers="keys", tablefmt="fancy_grid", showindex=False))
```

```python
import pandas as pd
import re
from tabulate import tabulate

# Load dataset
file_path = "final_mtn.csv"
df = pd.read_csv(file_path)

# Convert text to lowercase for uniformity
df["commentText"] = df["commentText"].astype(str).str.lower()

# Remove special characters and extra spaces
df["commentText"] = df["commentText"].apply(lambda x: re.sub(r'[^\w\s]', '', x).strip())

# Define keywords for MyMTN-related feedback
mymtn_keywords = [
    "mymtn", "mymtn app", "mtn app", "mtn application", "mymtn ng", "mtn nigeria app",
    "application", "mobile app", "mtn service app"
]

# Filter comments related to MyMTN app
df_mymtn = df[df["commentText"].str.contains(r'\b(' + '|'.join(mymtn_keywords) + r')\b', na=False, regex=True)]

# Further filter only strong positive comments (Threshold: > 0.8)
df_mymtn_positive = df_mymtn[df_mymtn["sentiment/positiveScore"] > 0.8]

# Select the top 5 most positive comments based on sentiment score & likes
df_top_positive_mymtn = df_mymtn_positive.nlargest(5, ["sentiment/positiveScore", "commentLikesCount"])[
    ["commentAuthor", "commentText", "sentiment/positiveScore", "commentLikesCount"]
]

# Display results in a formatted table
print("\n=== Top 5 Most Positive Comments for MyMTN App ===")
print(tabulate(df_top_positive_mymtn, headers="keys", tablefmt="fancy_grid", showindex=False))
```

```python
import pandas as pd
import re
from tabulate import tabulate

# Load dataset
file_path = "final_mtn.csv"
df = pd.read_csv(file_path)

# Preprocessing function: convert text to lowercase, remove punctuation, and extra spaces
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra whitespace
    return text

# Apply preprocessing to the commentText column
df["cleanedCommentText"] = df["commentText"].apply(preprocess_text)

# Define regex patterns for MyMTN-related feedback using word boundaries for accuracy
mymtn_patterns = [
    r"\bmymtn\b",           # matches "mymtn"
    r"\bmymtn app\b",       # matches "mymtn app"
    r"\bmtn app\b",         # matches "mtn app"
    r"\bmtn application\b", # matches "mtn application"
    r"\bmymtn ng\b"         # matches "mymtn ng"
]

# Function to check if a comment is related to MyMTN using the defined patterns
def is_mymtn_related(text):
    return any(re.search(pattern, text) for pattern in mymtn_patterns)

# Filter comments that are related to MyMTN/MTN app
df_mymtn = df[df["cleanedCommentText"].apply(is_mymtn_related)]

# Further filter only positive comments
df_mymtn_positive = df_mymtn[df_mymtn["sentiment/finalClassification"] == "positive"]

# Select the top 5 most positive comments based on sentiment/positiveScore
df_top_positive_mymtn = df_mymtn_positive.nlargest(5, "sentiment/positiveScore")[["commentAuthor", "commentText", "sentiment/positiveScore"]]

# Display results in a formatted table
print("\n=== Top 5 Most Positive Comments for MyMTN App ===")
print(tabulate(df_top_positive_mymtn, headers="keys", tablefmt="fancy_grid", showindex=False))
```
