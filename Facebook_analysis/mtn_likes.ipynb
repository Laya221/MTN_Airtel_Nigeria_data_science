```python

# Create a pivot table to count reactions for each post
result = df.pivot_table(
    index=['facebookId', 'facebookUrl'],  # Group by facebookId and facebookUrl
    columns='reaction',                  # Reaction types as columns
    aggfunc='size',                      # Count the number of occurrences
    fill_value=0                         # Fill missing values with 0
).reset_index()

# Ensure all reaction columns are present
all_reactions = ['angry', 'care', 'haha' ,'like', 'love', 'sad', 'wow']
for reaction in all_reactions:
    if reaction not in result.columns:
        result[reaction] = 0

# Reorder columns
result = result[['facebookId', 'facebookUrl'] + all_reactions]

# Save the result as a DataFrame
result_df = pd.DataFrame(result)

# Optional: Save to a CSV file
result_df.to_csv('mtn_reactions_summary.csv', index=False)
```

```python
# Find the post with the highest count for each reaction
highest_posts = {}
for reaction in ['angry', 'care', 'haha' ,'like', 'love', 'sad', 'wow']:
    max_row = result_df.loc[result_df[reaction].idxmax()]  # Get the row with the maximum value for the reaction
    highest_posts[reaction] = {
        'facebookId': max_row['facebookId'],
        'facebookUrl': max_row['facebookUrl'],
        'count': max_row[reaction]
    }

# Convert the result to a DataFrame
highest_posts_df = pd.DataFrame.from_dict(highest_posts, orient='index')

# Display the DataFrame
print(highest_posts_df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd

# Path to your ChromeDriver
chrome_driver_path = "chromedriver.exe"

# Set up Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-notifications")  # Disable Facebook notifications popup
driver = webdriver.Chrome(service=Service(chrome_driver_path), options=options)

# Facebook login credentials
email = "evetarek941@gmail.com"
password = " 221btarekeve"

# Log in to Facebook
def login_to_facebook():
    driver.get("https://www.facebook.com/")
    time.sleep(2)

    # Enter email
    email_box = driver.find_element(By.ID, "email")
    email_box.send_keys(email)

    # Enter password
    password_box = driver.find_element(By.ID, "pass")
    password_box.send_keys(password)

    # Click login button
    login_button = driver.find_element(By.NAME, "login")
    login_button.click()
    time.sleep(5)  # Wait for the page to load

# Scrape reactions from a Facebook post
def scrape_post_reactions(post_url):
    driver.get(post_url)
    time.sleep(5)  # Wait for the page to load

    # Hover over the reactions count to reveal details
    try:
        reactions_hover = driver.find_element(By.XPATH, "//span[contains(@aria-label, 'reactions')]")
        ActionChains(driver).move_to_element(reactions_hover).perform()
        time.sleep(3)  # Wait for reactions to load
    except Exception as e:
        print(f"Could not hover reactions: {e}")
        return None

    # Extract reaction counts
    reactions = {}
    try:
        reaction_elements = driver.find_elements(By.XPATH, "//div[contains(@aria-label, 'reaction')]")
        for element in reaction_elements:
            reaction_name = element.get_attribute("aria-label").split(":")[0].strip()
            reaction_count = int(element.get_attribute("aria-label").split(":")[1].strip().replace(",", ""))
            reactions[reaction_name] = reaction_count
    except Exception as e:
        print(f"Error extracting reactions: {e}")

    return reactions

# Main script
if __name__ == "__main__":
    # Log in to Facebook
    login_to_facebook()

    # List of Facebook post URLs to scrape
    post_urls = [
        "https://www.facebook.com/MTNLoaded/videos/1572063677204767/",
        # Add more post URLs here
    ]

    # Scrape reactions from each post
    data = []
    for url in post_urls:
        print(f"Scraping reactions for: {url}")
        reactions = scrape_post_reactions(url)
        if reactions:
            data.append({
                "post_url": url,
                **reactions
            })

    # Save data to a DataFrame
    df = pd.DataFrame(data)
    print(df)

    # Save the results to a CSV file
    df.to_csv("facebook_reactions.csv", index=False)

    # Close the browser
    driver.quit()
```
