```python
#Total posts per period

import matplotlib.pyplot as plt

def calculate_compression(original_size, compressed_size):
    compression_percentage = ((original_size - compressed_size) / original_size) * 100
    return compression_percentage

mtn_posts = 203
airtel_posts = 55

compression_percentage = calculate_compression(mtn_posts, airtel_posts)

# Create a pie chart
labels = ['MTN posts', 'Airtel posts']
sizes = [mtn_posts, airtel_posts]
explode = [0.1, 0]  # explode the MTN Posts slice
colors = ['yellow', 'red']  # Set colors for MTN and Airtel

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, colors=colors)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig('pie_chart.png', format='png', bbox_inches='tight')

plt.show()
```
